```python
import matplotlib.pyplot as plt
```

```python
x1=np.random.rand(500,1)
y1=np.random.rand(500,1)
```

```python
print(x1.shape)
print(y1.shape)
```

```python
plt.scatter(x1, y1)
plt.title("Scatter plot between $X$ and $Y$")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(x1, y1, c="purple",s=20)
plt.title("purple colored scatter plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.show()
```

```python
x2=2*x1
y2=4*y1

plt.scatter(x1,y1,c="green",s=20,label="from x1,y1")
plt.scatter(x2,y2, c="purple",s=40,label="from x2,y2")
plt.title("colored scatter plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.legend()
plt.show()
```

```python
plt.scatter(x1,y1,c="green",s=20,label="from x1,y1",marker="*",edgecolors="y")
plt.scatter(x2,y2, c="purple",s=40,label="from x2,y2",marker="o",edgecolors="b")
plt.title("Colored scatter plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.legend()
plt.show()
```

```python
plt.plot(x1, y1,c="g",linewidth=10)
plt.title("Line plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.show()
```

```python
plt.plot(x1,y1, c="red",label="From X1, Y1", linestyle="dotted")
plt.plot(x2,y2, c="green", label="From X2, Y2", linestyle="dashed")
plt.title("Labelled lines")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()
plt.show()
```

```python
plt.scatter(x1,y1, c="red",label="From X1, Y1")
plt.plot(x2,y2, c="green", label="From X2, Y2")
plt.title("Labelled lines")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x1,y1, c="red",label="Red Line")
plt.title("Red line")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(x1,y2, c="green", label="Green Line")
plt.title("Green line")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()

plt.subplots_adjust(left=2, right=4)
plt.show()
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=load_boston()
```

```python
X = data['data']
y = data['target']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
sample = X_test[1]
```

```python
lr.predict([sample])
```
